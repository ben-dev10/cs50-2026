## **Quiz**
Q: Which operator returns the memory address of a variable?
1) `*` (dereference operator) opposite of `&`, means go to that address
2) `$`
3) `&` ✅ (address of operator)
4) `->` (dereference and follow operator) works like a combination of both `*` and `&`

Q: Which SQL command is used to remove duplicate rows from a result set?
1) `REMOVE`
2) `UNIQUE` used when you want to find if keys are unique in your schema
3) `DISTINCT` ✅
4) `CLEAN`

Q: Where does malloc dynamically allocate memory from?
1) `HEAP` ✅
2) `STACK` 
3) `GLOBAL VARIABLES` 
4) `ASSEMBLY`

Q: Which of the following is NOT a built-in SQL feature to tackle race conditions?
1) `BEGIN TRANSACTION` 
2) `COMMIT` 
3) `ROLLBACK` 
4) `ENROLL` ✅
