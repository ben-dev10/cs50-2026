### **Loops**

````python
# cat.py

# while loop
i = 0;

while(i<3):
    print("meow")
    i+= 1 # i++ equivalent
````

````python
# cat.py

# for loop
for i in [0,1,2]:
    print("meow")

# explanation: for the first loop python will set i to 1 and execute the code inside, it does the same
# for 2, and then 3. after 3 since there is no value to set i to, the loop terminates
````

````python
# cat.py 

for _ in range(3):
    print("meow")

# explanation: python gives us a handy utility/function/data-type `range()` that would behave
# just as the list in the first example but is much smarter and simpler. Enter any value, or range
# of values (even skipping some numbers like listing all odd values for example) and python 
# returns those values to the loop. It starts from 0, so `range(3)` will return 0,1,2

# note: the use of the underscore as the variable name here just signals that it won't be used anywhere in
# the loop. `for-loop` in python requires a variable to be passed.
````

````python
# truncation: by default an operation that results in an infinitely repeating decimal points wii
# be truncated to a few decimal places

x = int(input("x: "))
y = int(input("y: "))
print(x/y)
````

````python
# floating-point imprecision: even if you expand to a higher decimal places to combat truncation,
# operations are still affected by imprecisions, due to memory constraints infinitely repeating decimal points (floats)
# will end up incorrect for higher floating points or decimal places, say 50 for example 

x = int(input("x: "))
y = int(input("y: "))

z = x/y
print(f"{z:.50}") # the syntax here `:.50` after the variable  means show up to 50 decimal places
````

### **Exceptions**
A way for functions to signal something went wrong with execution.

````python
# integer.py: without exceptions

n = input("Input: ")
if n.isnumeric():
    print("Integer")
else:
    print("Not integer")
````

````python
# integer.py: with exceptions

try:
    n = int(input("Integer: "))
    print("Integer")
except ValueError:
    print("Not an integer")
````

````python
scores = []
for i in range(3):
    score = int(input("Score: "))
    scores.append(score)

average = sum(scores) / len(scores)
print(f"Average: {average}")
````

````python
# phonebook.py

names = ["Kelly", "David", "Harvard"]
name = input("Name: ")

for n in names:
    if name == n:
        print("Found")
        break
else:
    print("Not found")


# note: One nice thing about python's for loop syntax is it accepts an else block, that is
# executed if no 'break' occurs in the main block.
````

````python
# phonebook.py

names = ["Kelly", "David", "Harvard"]
name = input("Name: ")

if name in names:
    print("Found")
else:
    print("Not found")
````

#### **Dicts**

````python
# phonebook2.py

people = {
    "Kelly": "+1-617-495-1000",
    "David": "+1-617-495-1000",
    "John": "+1-949-468-2750",
}

name = input("Name: ")

if name in people:
    number = people[name]
    print(f"Found: {number}")
else:
    print("Not found")
````

````python
# phonebook2.py

people = [
   {"name": "Kelly", "number": "+1-617-495-1000"},
   {"name": "David", "number": "+1-617-495-1000"},
   {"name":"John", "number": "+1-949-468-2750"}
]

name = input("Name: ")

for person in people:
    if person["name"] == name:
        number = person["number"]
        print(f"found: {number}")
        break
else:
    print("Not found")
````

### **Internal Libraries**

- `sys`: 
    - Give access to command-line arguments.
    - `argv`: is the list of arguments the user has typed

````python
# greet.py

from sys import argv

if len(argv) == 2:
    name = argv[1]
    print(f"Hello {name}")
else: 
    print("Hello World")

# explanation: if the user runs this program from the command line with `python agree.py`, return 
# "Hello World" if there is no argument, else output "Hello + first-argument"
````

````python
import sys

if len(sys.argv) !=2:
    print("Missing command-line arguments")
    sys.exit(1)

print(f"Hello {sys.argv[1]}")
sys.exit(0)

# note: python agree.py   = 1 argument
#       python agree.py David  = 2 command-line arguments
````
