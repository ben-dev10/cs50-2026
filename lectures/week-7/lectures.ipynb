### **CSV Operations With Python**


- python `csv` module: 
    - a CSV reader module used for CSV file operations.
    - the reader is stateful, it remembers where it is in memory.

````python
# favourites.py

import csv

with open("favourites.csv","r") as file:
    reader = csv.reader(file) # returns data in a list structure

    next(reader) # go to the next/second line in the file, after the header line

    for row in reader:
        favorite = row[1] # get index
        print(favorite)
````

````python
# favourites2.py: safeguard against column mis-match with `DictReader()`

import csv

with open("favourites.csv","r") as file:
    reader = csv.DictReader(file) # return data in a dictionary data structure

    for row in reader:
        favorite = row["language"] # get language property
        print(favorite)
````

````python
# favourites3.py

import csv

with open("favourites.csv","r") as file:
    reader = csv.DictReader(file) 
    
    scratch, c, python = 0, 0, 0

    for row in reader:
        favorite = row["language"] 
        
        if favorite == "Scratch":
            scratch += 1
        elif favorite == "C":
            c += 1
        elif favorite == "Python":
            python += 1

print(f"Scratch: {scratch}")
print(f"C: {c}")
print(f"Python: {python}")
````

````python
# favourites3b.py: a more optimized solution that can handle any amounts of
# favourites without having to re-edit the code

import csv

with open("favourites.csv","r") as file:
    reader = csv.DictReader(file) 
    
    counts = {}

    for row in reader:
        favorite = row["language"] 
        
        if favorite not in counts: # check if favorite exists in dict, which is empty initially
            counts[favorite] = 0 # add any new favorite you see as key and set it's initial value to 0
        
        counts[favorite] += 1 # increment key value by 1 on subsequent occurrences

for favorite in sorted(counts, key=counts.get, reverse=True):
    print(favorite, counts[favorite])
````

> You can sort dictionaries using the `sort(dictionary)` method. Leave as is to sort by key, alphabetically, `sort(dictionary, key=counts.get)` to sort by values, and a third parameter `reverse=True` to reverse the current sort.

````python
# favourites3c.py: KeyError Management

import csv

with open("favourites.csv","r") as file:
    reader = csv.DictReader(file) 
    
    counts = {}

    for row in reader:
        favorite = row["language"] 
        try:
            counts[favorite] += 1
        except KeyError:
            counts[favorite] = 1


for favorite in counts:
    print(favorite, counts[favorite])
````

### **SQLITE**
A simple lightweight variant of MYSQL. SQLITE provides a CLI tool that can be used to manage its databases. Install it on your system, and run `sqlite3` to launch it.

```bash
$ sqlite3
# SQLite version 3.51.0 2025-11-04 19:38:17
# Enter ".help" for usage hints.
# Connected to a transient in-memory database.
# sqlite> 
```


**CLI Commands:**
- `.schema`: 
    outputs the structure of a table, as a command that can be used to generate that table  
    **Syntax**: `.schema table`

- `.mode`
- `.import`
- `.quite`

#### **Basic Workflow**

[1] Create a new database with:
```bash
sqlite3 favourites.db  # it creates a binary db file into the current folder and opens the
# interactive interpreter
```

[2] Import data source (csv,...or any suitable format)
```bash
sqlite> .mode csv  # change mode to csv, necessary for importing csv data
sqlite> .import favourites.csv favourites # favourites is the table name
sqlite> .quit
```

[3] Now anytime you want work with this database, simply pass the db to the `sqlite command`
```bash
sqlite3 favourites.db  # favourites.db now exists so no setup required
# SQLite version 3.51.0 2025-11-04 19:38:17
# Enter ".help" for usage hints.

# use .schema to see the shape (rows and columns) of the data structure
sqlite> .schema
# CREATE TABLE IF NOT EXISTS "favourites"(
# "Timestamp" TEXT, "language" TEXT, "problem" TEXT);
```

````python
-- find how many people chose 'C'
SELECT COUNT(*) FROM favourites WHERE language = 'C';

-- find how many people chose 'C' and a given problem
SELECT COUNT(*) FROM favourites WHERE language = 'C' AND problem = 'This is too easy';

-- find how many people chose 'C' and a given problem 2
SELECT COUNT(*) FROM favourites WHERE language = 'C' AND problem = 'Hello it''s me'; -- use double single quotes to escape a single quote
SELECT COUNT(*) FROM favourites WHERE language = 'C' AND (problem = 'Hello it''s me' OR problem = 'Hello, it''s Me'); 
SELECT COUNT(*) FROM favourites WHERE language = 'C' AND (problem = 'Hello it''s me' OR problem = 'Hello, it''s Me'); 

-- fuzzy/wildcard search
SELECT COUNT(*) FROM favourites WHERE language = 'C' AND problem LIKE 'This is %'; 

-- Group by
SELECT language, COUNT(*) FROM favourites GROUP BY language;

-- Order results
SELECT language, COUNT(*) FROM favourites GROUP BY language ORDER BY COUNT(*) DESC; -- 1
SELECT language, COUNT(*) AS n FROM favourites GROUP BY language ORDER BY n DESC; -- 2: Setting alias (COUNT(*) as n)
SELECT language, COUNT(*) AS n FROM favourites GROUP BY language ORDER BY n DESC LIMIT 1; -- with limit (limit result to only 1 row)
````

#### **Data Relations**
Relations involve linking data across tables using _**IDs (keys)**_. Relations between these tables can be any of these forms:
- `One-to-one`: where one row in one table maps to one row in another table 
- `One-ton-many`


**Keys**
- `PRIMARY KEY`: identifies rows from the current table
- `FOREIGN KEY`: identifies rows for a foreign table

**Schemas**
```sql
>>> .schema shows
CREATE TABLE shows {
    id INTEGER,
    title TEXT NOT NULL,
    year NUMERIC,
    episodes INTEGER,
    PRIMARY KEY(id)
}

>>> .schema genres
CREATE TABLE genres {
    show_id INTEGER NOT NULL,
    genre TEXT NOT NULL,
    FOREIGN KEY(show_id) REFERENCES shows(id) -- a foreign key "show_id" that references
    -- "id" key in shows table 
};

>>> .schema ratings
CREATE TABLE ratings{
    show_id INTEGER NOT NULL UNIQUE,
    rating REAL NOT NULL,
    votes INTEGER NOT NULL,
    FOREIGN KEY(show_id) REFERENCES shows(id)
}
```

> **Naming conventions for keys(primary & foreign)**:   
> primary: `id`
> foreign: `table-name_key-name` => `show_id`

**Selections with IDs**

````python
SELECT * FROM ratings WHERE rating >= 6.0 LIMIT 10; 

SELECT * FROM genres LIMIT 10;
SELECT * FROM shows WHERE id = 5;

-- the statement in parenthesis is executed first and is  what queries the ratings table using its show_id
SELECT * FROM shows WHERE id IN 
( SELECT show_id FROM ratings WHERE rating>= 6.0 ) LIMIT 10;

SELECT title FROM shows WHERE id IN 
( SELECT show_id FROM ratings WHERE rating>= 6.0 ) LIMIT 10;
````

**`JOINS`**:  
When you want to display values from 2 or more tables into one single value/row-matching table, where the unique IDs should line up.

````python
SELECT * FROM shows
JOIN ratings ON shows.id = ratings.show_id
WHERE rating >= 6.0 LIMIT 10;

SELECT title, rating FROM shows
JOIN ratings ON shows.id = ratings.show_id
WHERE rating >= 6.0 LIMIT 10;
````

**One-to-many**  
**Example**: One movie can belong to one or more genres

````python
SELECT genre FROM genres WHERE show_id = 5; -- get genres for entry/row with id = 5
SELECT genre FROM genres WHERE show_id = ( -- nested query syntax
    SELECT id FROM shows WHERE title = 'Hidden Universe'
) -- get genres using title instead

-- JOINS with one to many
SELECT title, genre FROM shows
JOIN genres ON shows.id = genres.show_id
WHERE id = 15; -- show a joined table of a shows table and its genre

SELECT genre FROM shows
JOIN genres ON shows.id = genres.show_id
WHERE id = 15; -- join a given movie and its genre into a table but only show the genre column 

SELECT * FROM shows WHERE title LIKE 'The %' AND year = '2005'; 
````

**`Many-to-Many`**

````python
SELECT * FROM shows WHERE title = 'The Window' AND year = 2008; 
SELECT id FROM shows WHERE title = 'The Window' AND year = 2008;

SELECT person_id FROM stars WHERE show_id = (
    SELECT id FROM shows WHERE title = 'The Window' AND year = 2008
);

-- selects all stars(actors) staring in this show
SELECT name FROM people WHERE id IN ( -- get names of all people selected below based on ids
     SELECT person_id FROM stars WHERE show_id = ( -- gets all related person_ids
        SELECT id FROM shows WHERE title = 'The Window' AND year = 2008 -- gets show_id
     )
);


SELECT * FROM people WHERE name LIKE 'Nancy %';
-- 1a: get show_id (# of shows) for actress Nancy Wang
SELECT show_id FROM stars WHERE person_id = ( 
    SELECT id FROM people WHERE name = 'Nancy Wang'
);
-- 1b: get title using show_ids for actress Nancy Wang
SELECT TITLE FROM shows WHERE id IN (
    SELECT show_id FROM stars WHERE person_id = ( 
        SELECT id FROM people WHERE name = 'Nancy Wang'
    )
);
````

In the queries above, which involved querying 2 different tables for values, nested queries were used (use of parenthesis to group related queries), but this is something `JOINS` were made to solve.

````python
-- NOTE: duplicates are possible with joins --
-- 1c: get title using show_ids for actress Nancy Wang
SELECT title FROM shows
JOIN stars ON shows.id = stars.show_id
JOIN people ON stars.person_id = people.id
WHERE name = 'Nancy Wang';

-- 1d: get title using show_ids for actress Nancy Wang
SELECT title FROM shows, stars, people
WHERE shows.id = stars.show_id
AND people.id = stars.person_id
AND name = 'Nancy Wang';



-- joins shows, people and stars tables into one giant table
SELECT * FROM shows
JOIN stars ON shows.id = stars.show_id
JOIN people ON stars.person_id = people.id;
````

> **Note**: 
> - `ratings` table was `One-to-One`, `genres` was `One-to-many`, `stars` is `Many-to-many`
> - `ORMs` can be used to further abstract away SQL queries and relational database management in general.

#### **Indexes**
Optimize operations on larger datasets or frequently used queries using `INDEX`.

````python
-- use timer to monitor query speeds
sqlite3> .timer ON
sqlite3> SELECT * FROM shows WHERE title LIKE 'The %'; -- 2a
--╭──────┬─────────────────────────┬──────┬──────────╮
--│  id  │          title          │ year │ episodes │
--╞══════╪═════════════════════════╪══════╪══════════╡
--│    1 │ The Hand                │ 2008 │       19 │
--│ 1197 │ The Light: Unleashed    │ 2016 │       14 │
--│ 1198 │ The Station: Forever    │ 2013 │       83 │
--╰──────┴─────────────────────────┴──────┴──────────╯
--Run Time: real 0.392972 user 0.000000 sys 0.031250
````

`CREATE INDEX name ON table (column,...);`

````python
CREATE INDEX title_index ON shows (title); -- create an index called title_index for the 
-- shows table on title column

-- Rerun query 2a above and should now get somewhat higher speeds (depends)
SELECT * FROM shows WHERE title LIKE 'The %';
-- Run Time: real 0.340516 user 0.000000 sys 0.046875
````

But what really are indexes and how do they work? By default SQLITE's database query runs on a linear algorithmic scale which is quite fast for most use cases. But when you opt for an index for a particular column(s) for a given table, the SQLITE engine restructures the data for that column into a different data structure, `B-tree`, that enables algorithms for queries to be completed in logarithmic time.

#### **SQL Injection Attacks**
In our previous example in (favourites.py: 4b) our use of f-strings in the query means we are blindly trusting the user to type the right characters everytime which might not always be the case. A simple input like `'malan david'--` can effectively cut short the rest of the query to the right and just return the user data without any checks. This is an example of an SQL injection attack.

So the advice here is to never interpolate values this way but to always use a dedicated library that would sanitize the inputs for you.

````python
# favourites.py: 4c
from cs50 import SQL

db = SQL("sqlite:///favourites.db")

favorite = input("Favourites: ")

# the question mark ? is used as a placeholder for the second favorite parameter, that would be sanitized before
# injected into the query string
rows = db.execute(f"SELECT COUNT(*) AS n FROM favourites WHERE problem = ?", favorite) 

row = rows[0]

print(row["n"])
````
