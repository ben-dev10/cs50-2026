## **Internet**
Internet is a network (and network of networks, locally or publicly, wired or wirelessly) of computers.

- Routers: routes data(packets) among computers

### **[a] Web Protocols: TCP/IP**
#### **Internet Protocol (IP):**   
used for assigning unique identifiers to every device connected to the internet. Consists of a string of numbers of four sections separated by a period.
$$\#.\#.\#.\#$$
$$Eg: 10.234.2.173$$

$\#$ = a number between 0 - 255. 

For the **IPv4** string, each number is stored with 8 bits, so each number * 4, equals **32bits** in total for an IP address. However there is a problem with this because now you can only count up to 2^(32) approximately 4billon, which means only 4billion devices can be identified.

**1Pv6** uses 128bits for each number with gives a wide range numbers enough for the current age and in theory should be so for the next few decades or centuries.  


#### **TCP**:  
Allows a computer to distinguish the type of package it has received i.e email, video ..etc. In simple terms it guarantees delivery.

##### **TCP Header Format**:
- Ports: source port, destination port
- Sequence number
- Acknowledgement number
- Data offset
- Checksum
- Urgent pointer, Options, Padding,
- **Data: actual data you send**

<br />

**[1] Ports:**  
Numeric values from 0 and below 1,000. The are like entry points for specific applications, and so if another computer sends a piece of data and labels it as port 80, your computers understands that data is a webpage and not something else as port 80 is used for HTTP requests or activities.

**Examples of ports:**
- 80:  HTTP
- 443: HTTPS
...


**[2] Sequence numbers:**  
When a file (usually big files) is being transferred over a network it is usually broken up into hundreds of pieces or packets, each packet is assigned a 32bit number that is used to re-construct the data at the receiving node, this number is the **sequence number**

#### **DHCP (Dynamic Host Configuration Protocol)**
A server responsible for assigning IPs to devices. On bootup or reboot, your device (mobile, router or even DNS servers) often broadcasts a message asking for an IP address, a DHCP server responds and gives your device it's IP.

#### **Internet Datagram Headers**
A standard structure and components of bits that makes up a packet of data. 


#### **DNS: Domain Name System**
A computer used to correlate "domain names" or "fully qualified domain names" as they are technically called ("google.com", "apple.com") to corresponding "IP addresses". 


DNS servers are installed in many machines in the network and exists in herarchical order, from **Root DNS servers** → .. → Router → ... → PC, that a computer can query for an IP address when needed. If aa domain can't be found in a node lower down in the hierarchy, it is escalated until it is eventually cached from the root servers.

#### **Internet Registrars**
Companies that provides domain name services, through which you can register to get you very own personal email and they handle DNS management for you.

#### **Regular Expressions (regexes)**  
Used for pattern matching in an input string.

Docs: https://developer.mozilla.org/en-US/docs/Web/javaScript/Guide/Regular_expressions

**Syntax**:
- `.`: any single character (except line characters)
- `*`: zero or more times
- `+`: one or more times
- `?`: zero or 1 time
- `{n}`: n occurrences
- `{n,m}`: at least n occurrences, at most m occurrences
- `[0123456789]`: any one of the enclosed characters
- `[0-9]`: any one of the range of characters
- `\d`: any digit
- `\D`: any character that is not a digit
- ...

**Example:**  
```html
pattern=".+@.+\.edu" (valid match: "malan@havard.edu")
```
`.`: matches any text before the @ symbol, plus(+) `@`: at symbol must be present plus(+) `.`: any text, plus(+) `\.`: a literal dot, escaped with a slash, then `edu`: an "edu" string at the end.

````python
// attach a submit listener to all form elements and show an alert on submit using value from 
// element '#name'
document.querySelector('form').addEventListener('submit', function(event){
    alert('hello, ' + document.querySelector('#name').value);
    event.preventDefault();
})
````

````python
// wrap your code with this snippet below to listen for full page load (or whole DOM is loaded into memory) 
// before executing the contents within
document.addEventListener('DOMContentLoaded', function(){
    // some code
})
````

````python
document.addEventListener('DOMContentLoaded', function(){
    let input = document.querySelector('input');
    input.addEventListener('keyup', function(event){
        let name = document.querySelector('p');

        if(input.value){
            name.innerHTML = `hello, ${input.value}`;
        }else{
            name.innerHTML = 'hello, whoever you are';
        }
    })
})
````

````python
// blink.js

function blink(){
    let body = document.querySelector('body');

    if(body.style.visibility == 'hidden'){
        body.style.visibility = 'visible';
    }else{
        body.style.visibility = 'hidden';
    }
}

window.setInterval(blink, 500)  // call the blink function every 500ms
````

````python
// geolocation.js

navigator.geolocation.getCurrentPosition(function(position){
    document.querySelector("#location").innerHTML = position.coords.latitude+ ', ' + position.coords.longitude;
})
````
