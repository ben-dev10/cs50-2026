### **Pointers**
A pointer is a $variable$ that can store a memory address...then the compiler and computer decide where to put that value in memory.

````c
#include <stdio.h>

int main(void)
{
    int n = 50;
    int *p = &n; // a pointer p (*p) storing the addresses of variable n (&n)

    printf("%i \n", n); // variable
    printf("%p \n", p); // pointer to variable address
}
````

- $p$ (pointers) is usually 64 bits (8bytes) long
- computers do this (store a pointer alongside a given value) all the time under the hood but C lets us see where the variables are stored in memory and can allow us to store values at any address we choose.

````c
#include <stdio.h>

int main(void)
{
    int n = 50;
    int *p = &n; // a pointer p (*p) storing the addresses of variable n (&n)

    printf("%i \n", *p); // print out a value (50) using a pointer (i.e a known address) *p
    // *p is also know as dereference p
}
````

### **String Addresses**
You may think of a string as an array of characters which is true. But given a string `s = "HI!"` in memory, the computer only stores the memory address (pointer) of the first character "H", this works because the entire string is stored in a contiguous fashion appended with the NUL character `\O` which tells the computer the end of that string.

````c
#include <stdio.h>
#include <cs50.h>
 
int main(void)
{
    string s = "HI!";

    printf("%p \n", &s); // entire string
    printf("%p \n", &s[0]); // first char, contiguous from here
    printf("%p \n", &s[1]);
    printf("%p \n", &s[2]);
    printf("%p \n", &s[3]);  // NUL terminator (\0)
}
````

````c
// Note: the "string" comes form <cs50.h>, actual string type below
#include <stdio.h>

typedef char *string;

int main(void)
{
    char *s = "HI!"; // OR: string s = "HI!";

    printf("%p \n", &s); // entire string
    printf("%p \n", &s[0]); // first char, contiguous from here
    printf("%p \n", &s[1]);
    printf("%p \n", &s[2]);
    printf("%p \n", &s[3]);  // NUL terminator (\0)
}

// internal implementation of string is as follows

````

Using **pointer arithmetics** as an alternative to accessing array (string) items.

````c
#include <stdio.h>

#include <stdio.h>
 
int main(void)
{
    char *s = "HI!";
    // using the normal string[index] notation
   printf("%c \n", s[0]);
   printf("%c \n", s[1]);
   printf("%c \n", s[2]);

   printf("---------- \n");

   // using 
   printf("%c \n", *s);  // %i will give the ASCII value, %c will give the character
   printf("%c \n", *(s + (1)));  // first string address + 1, i.e next item
   printf("%c \n", *(s + (2)));  
   printf("%c \n", *(s + (3)));  
}
````

Print out a sequence of string characters

````c
#include <stdio.h>

#include <stdio.h>
 
int main(void)
{
    char *s = "HI!";
   printf("%s \n", s); // print out starting at the first address
   printf("%s \n", s + 1); // start from 1 place after the first address
   printf("%s \n", s + 2); // start from 2 places after the first address
   printf("%s \n", s + 3); // prints the NUL character which is empty space

   // this works because of the NUL terminator indicating to printf where to stop
}
````

### **Copying & Memory Allocation**

Below does a shallow copy by creating a new pointer t that points to s as a shallow copy.

Running this will result in an error (Segmentation fault, exit code -11), because we are directly trying to modify `t` (`toupper`) but its guarding against that because its a copy in itself thus pointing to a read-only memory at `s`. Use `get_string` instead.

````c
//%cflags:-lcs50
#include <stdio.h>
#include <ctype.h>
#include <string.h>
#include <cs50.h>
 
int main(void)
{
    char *s = "Hello"; // char *s = get_string("s: ");
    char *t = s;

    t[0] = toupper(t[0]);

   printf("s: %s\n", s);
   printf("t: %s\n", t);
}
````

##### **`malloc()` and `free` - deep copying**
`malloc` (or memory allocation) returns the address of a chunk of memory

Other useful methods to be used
- `strcopy()`


````c
#include <stdio.h>
#include <ctype.h>
#include <string.h>
#include <cs50.h>
#include <stdlib.h>
 
int main(void)
{
    char *s = "hello";
    char *t = malloc(strlen(s)+1); // you can hardcode a number but that is error prone so do 
    // that dynamically with strlen() +1 means include the null character you can also 
    // hardcode the null character with: t[4] = '\0'; but this too is error prone, use the first.

    // manual copying of each characters
    // for(int i = 0, n = strlen(s); i <= n ; i++){
    //     t[i] =  s[i];
    // }

    // using strcpy(destination, source)
    strcpy(t,s);

    t[0] = toupper(t[0]); // turn first character of t to uppercase;

    printf("s: %s\n", s);
    printf("t: %s\n", t);
}
````

#### **Error checks**
Up until now we assumed that since we got the right responses everything might be fine. But it isn't, we have failed to check for edge cases in our programs. 

Below is an updated program from before we better checks in place.

````c
//%cflags:-lcs50
#include <stdio.h>
#include <ctype.h>
#include <string.h>
#include <cs50.h>
#include <stdlib.h>
 
int main(void)
{
    char *s = get_string("s: ");
    
    if(s == NULL){
        return 1; // strlen can fail, say a very large input string for example, 
        // guard against that by checking for a NULL return from strlen
    }

    char *t = malloc(strlen(s)+1); 
    if(t == NULL){
        return 1; // same as get_string condition
    } 

    strcpy(t,s);

    if(strlen(s) > 0){
        t[0] = toupper(t[0]); // check if the user provides a value before converting to uppercase
    }
    
    printf("s: %s\n", s);
    printf("t: %s\n", t);

    free(t); // if you asked the computer for memory, remember to give it back at the end of your program,
    // lest you cause a memory leak.
}
````
