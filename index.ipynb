### **Decimal vs Binary**

**Decimal (base 10)**:   
$123$   
    = $1(10^2) + 2(10^1) + 3(10^0)$  
    $= 100 + 20 + 1$

**Decimal → Binary (base 2)**:   
$123$   
    = $1(10^2) + 2(10^1) + 3(10^0)$  
    $= 100 + 20 + 1$

**Binary → Decimal (base 2)**:   
$11111111$  
    = $1(2^7) + 1(2^6) + 1(2^5) + 1(2^4)+ 1(2^3) + 1(2^2)+ 1(2^1)+ 1(2^0)$  
   $= 128 + 64 + 32 + 16 + 8 + 4 + 2 + 1 = 255$


> **Note**: Base exponents go from `0` on the last number to the right and increases by one towards the left.

- `1 bit` can store up to `2`
- `2 bits` can store up to `4`
- `3 bits` can store up to `8`
- `4 bits` can store up to `16`  
...  
- `8 bits` can store up to `256`
