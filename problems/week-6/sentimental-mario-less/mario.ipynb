```python
# version 1: doesn't handle strings

def get_int(prompt):
    # ensure input is only integer
    integer = int(input(prompt))

    # re-prompt until correct response  
    while (integer <= 0 or integer > 8 or type(integer) == str):
        integer = int(input(prompt))
    
    return integer


# right-angled triangle
def build_pyramid(height):
    for i in range(height):
        # print spaces
        print(" " * (height - i), end="")
        print(("#"* (i+1)))


# Get height
height = get_int("Height: ")
build_pyramid(height)
```
