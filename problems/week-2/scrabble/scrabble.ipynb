# Scrabble (Problem 2.1)

In the game of Scrabble, players create words to score points, and the number of points is the sum of the point values of each letter in the word.

For example, if we wanted to score the word “CODE”, we would note that the ‘C’ is worth 3 points, the ‘O’ is worth 1 point, the ‘D’ is worth 2 points, and the ‘E’ is worth 1 point. Summing these, we get that “CODE” is worth 7 points.

In a file called `scrabble.c` in a folder called `scrabble`, implement a program in C that determines the winner of a short Scrabble-like game. Your program should prompt for input twice: once for “Player 1” to input their word and once for “Player 2” to input their word. Then, depending on which player scores the most points, your program should either print “Player 1 wins!”, “Player 2 wins!”, or “Tie!” (in the event the two players score equal points).

### **Solution 1**

```c
#include <stdio.h>
#include <cs50.h>
#include <string.h>
#include <ctype.h>

int POINTS[] = {1, 3, 3, 2, 1, 4, 2, 4, 1, 8, 5, 1, 3, 1, 1, 3, 10, 1, 1, 1, 1, 4, 4, 8, 4, 10};

int compute_score(string word);

int main(void)
{

    // ask user for 2 inputs
    string player1 = "Hello"; // player1 = get_string("Player 1: "); 
    string player2 = "World"; // get_string("Player 2: ");

    // compute score
    int score1 = compute_score(player1);
    int score2 = compute_score(player2);

    if (score1 > score2)
    {
        printf("Player 1 wins! \n");
    }
    else if (score1 < score2)
    {
        printf("Player 2 wins! \n");
    }
    else
    {
        printf("Tie!\n");
    }
}

int compute_score(string word)
{
    int score = 0;

    for (int i = 0, len = strlen(word); i < len; i++)
    {
        if (isupper(word[i]))
        {
            score += POINTS[word[i] - 'A'];
        }
        else if (islower(word[i]))
        {
            score += POINTS[word[i] - 'a'];
        }
    }

    return score;
}
```

### **Solution 2**

Solution 2 is more verbose and frankly, inefficient but it demonstrates an alternative working solution. I recommend solution 1 over this.

```c
#include <stdio.h>
#include <cs50.h>
#include <string.h>
#include <ctype.h>

typedef struct
{
    char letter;
    int score;
} ScoreTable;

int main(void)
{
    ScoreTable Scores[] = {
        {'A', 1}, {'B', 3}, {'C', 3}, {'D', 2}, {'E', 1},
        {'F', 4}, {'G', 2}, {'H', 4}, {'I', 1}, {'J', 8},
        {'K', 5}, {'L', 1}, {'M', 3}, {'N', 1}, {'O', 1},
        {'P', 3}, {'Q', 10}, {'R', 1}, {'S', 1}, {'T', 1},
        {'U', 1}, {'V', 4}, {'W', 4}, {'X', 8}, {'Y', 4},
        {'Z', 10}
    };

     string player1 = "Hello"; // player1 = get_string("Player 1: "); 
    string player2 = "World"; // get_string("Player 2: ");

    int score1 = 0;
    int score2 = 0;

    int table_size = sizeof(Scores) / sizeof(Scores[0]);

    // Score Player 1
    for (int i = 0; i < strlen(player1); i++)
    {
        char letter = toupper(player1[i]);

        for (int j = 0; j < table_size; j++)
        {
            if (letter == Scores[j].letter)
            {
                score1 += Scores[j].score;
                break;
            }
        }
    }

    // Score Player 2
    for (int i = 0; i < strlen(player2); i++)
    {
        char letter = toupper(player2[i]);

        for (int j = 0; j < table_size; j++)
        {
            if (letter == Scores[j].letter)
            {
                score2 += Scores[j].score;
                break;
            }
        }
    }

    // Determine winner
    if (score1 > score2)
    {
        printf("Player 1 wins!\n");
    }
    else if (score2 > score1)
    {
        printf("Player 2 wins!\n");
    }
    else
    {
        printf("Tie!\n");
    }
}
```
