Comparing strings with `strcmp` (`string.h`).

```c
//%cflags:-lcs50
#include <stdio.h>
#include <string.h>
#include <cs50.h>
 
int main(void)
{
    string s = "Hello"; 
    string t =  "Hello"; 

    if(strcmp(s,t) == 0){
        printf("Same \n");
    }else{
        printf("Different \n");
    }
}
```

```c
#include <stdio.h>
 
int main(void)
{
    char *t = "Hi";
    
    
   printf("%c \n", t[0]);
   printf("%c \n", *t);
   // first character if t is equal to the dereference(point to values) of t
}
```
